**KNN (ML with scikit_learn)** 3V-codes

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cancer = load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify= cancer.target, random_state = 42)
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
```

```python
print('Accuracy of KNN n=5, on Training set:{:.3f}'.format(knn.score(X_train, y_train)))
print('Accuracy of KNN n=5, on test set:{:.3f}'.format(knn.score(X_test, y_test)))
```

```python
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify= cancer.target, random_state = 66)

training_accuracy = []
test_accuracy = []

neighbors_settings = range(1, 11)

for n_neighbors in neighbors_settings:
  clf = KNeighborsClassifier(n_neighbors=n_neighbors)
  clf.fit(X_train, y_train)
  training_accuracy.append(clf.score(X_train, y_train))
  test_accuracy.append(clf.score(X_test, y_test))
```

```python
plt.plot(neighbors_settings, training_accuracy, label = 'Acc of training set' )

plt.plot(neighbors_settings, test_accuracy, label = 'Acc of test set')

plt.ylabel("Accuracy")
plt.xlabel("No of Neighbors")

plt.legend()
```

**Random Forest Algorithm** (ML with scikit_learn) - 3V codes


---



```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()

P_train, P_test, Q_train, Q_test = train_test_split(cancer.data, cancer.target, random_state = 0)

forest = RandomForestClassifier(n_estimators=100, random_state = 0)

forest.fit(P_train, Q_train)
```

```python
print('Accuracy od Training Subset:{:.3f}'.format(forest.score(P_train, Q_train)))
print('Accuracy od Test Subset:{:.3f}'.format(forest.score(P_test, Q_test)))
```

```python
import matplotlib.pyplot as plt

import numpy as np
%matplotlib inline
```

```python
n_features = cancer.data.shape[1]

plt.barh(range(n_features), forest.feature_importances_, align = 'center')
plt.yticks(np.arange(n_features), cancer.feature_names)

plt.xlabel('Feature Importance')
plt.ylabel("Feature")

plt.show()
```

**Potential Advantage of Random Forest**


1.   Powerful and widely implemented
2.   perform well with default settings
3. dont require scalling of the data
4. randomization makes them better than single Decision tree

**Parameter to Tune**

* n_jobs - no. of cores to use for training (n_jobs = 1, for all cores)
* n_estimators - how many tree to use (more is always better)

* max_depth, for pre-pruning
* max_features, for randomization
        * max_features = sqrt(n_features), for classification 
        * max_features = log2(n_features), for regression
* etc;

**Potential Disadvantages od Random Forests**

* not so good performance on very high dimensional and sparse data(text data)

* large datasets require more resources for training (time, CPU's etc)

* cannot be visualize as well as single Decision Tree

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()

M_train, M_test, N_train, N_test = train_test_split(cancer.data, cancer.target, random_state = 0)

mlp = MLPClassifier(random_state = 42)

mlp.fit(M_train, N_train)
```

```python
print('Accuracy od Training Subset:{:.3f}'.format(mlp.score(W_train_scaled, N_train)))
print('Accuracy od Test Subset:{:.3f}'.format(mlp.score(W_test_scaled, N_test)))
```

```python
print('Accuracy od Training Subset:{:.3f}'.format(mlp.score(W_train_scaled, N_train)))
print('Accuracy od Test Subset:{:.3f}'.format(mlp.score(W_test_scaled, N_test)))
```

**Support Vector Machines (SVMs) in scikit_learn** 

* for classification and regression(SVCs, SVRs)
* can be applied on linear and non-linear model
* look for the best separating line or decision boundary
* look for the largest margin

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state = 0)

svm = SVC()

svm.fit(X_train, y_train)
```

```python
print('Accuracy of Training Subset:{:.3f}'.format(svm.score(X_train, y_train)))
print('Accuracy of Test Subset:{:.3f}'.format(svm.score(X_test, y_test)))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(X_train.min(axis = 0), 'o', label = 'Min')
plt.plot(X_train.max(axis = 0),'v', label = 'Max')
plt.xlabel("feature index")
plt.ylabel('feature magnitude in log scale')
plt.yscale('log')
plt.legend(loc = "upper right")
```

```python
min_train  = X_train.min(axis = 0)
range_train = (X_train - min_train).max(axis = 0)

X_train_scaled = (X_train - min_train) / range_train

print('minimum per feature \n {}'.format(X_train_scaled.min(axis = 0)))
print('maximum per feature \n {}'.format(X_train_scaled.max(axis = 0)))
```

```python
X_test_scaled = (X_test - min_train) / range_train

svm = SVC()
svm.fit(X_train_scaled, y_train)
```

```python
print('Accuracy of Training Subset:{:.3f}'.format(svm.score(X_train_scaled, y_train)))
print('Accuracy of Test Subset:{:.3f}'.format(svm.score(X_test_scaled, y_test)))
```

```python
svm = SVC(C = 1000)
svm.fit(X_train_scaled, y_train)

```

```python
print('Accuracy of Training Subset:{:.3f}'.format(svm.score(X_train_scaled, y_train)))
print('Accuracy of Test Subset:{:.3f}'.format(svm.score(X_test_scaled, y_test)))   #overfitting occurs 
```

**Advantages And Disadvantages of SVMs (Scikit-learn)**

**Stronger Points:**
* they are versatile
* can build complex decision boundaries on low-dimensional data
* can work well on high-dimensional data with relatively small sample size
*etc..

**Weaker Points:**

* don't perform well on high-dimensional data with many samples (i.e > 100k)
* preprocessing may be required => implies knowledge and understanding of hyper-parameters
* harder to inspect and visualize
* etc...

**Alternatives :**

* DT and Random Forest (required less/no preprocessing of data, easier to understand, inspect and visualize)


**Good Practice:**

* data scaling
* other pre-processing
*choosing an appropriate kernal
* tuning hyper-parameter: C, gamma, etc.
