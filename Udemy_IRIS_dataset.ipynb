**Uncertainty Estimation For Multi-CLass Datasets**

```python
from sklearn.datasets import load_iris
```

```python
from sklearn.model_selection import train_test_split

from sklearn.ensemble import GradientBoostingClassifier
```

```python
iris = load_iris()
```

```python
iris.target_names
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, random_state=42)
```

```python
from scipy.sparse.construct import random
gbrt = GradientBoostingClassifier(learning_rate = 0.01, random_state = 0)
```

```python
gbrt.fit(X_train, y_train)
```

```python
print('the decision function for the 3- class iris dataset:\n\n {}'.format(gbrt.decision_function(X_test)[:20]))
```

```python
print("predicted probability for the sample in the iris dataset:\n\n{}".format(gbrt.predict_proba(X_test)[:10]))
```
